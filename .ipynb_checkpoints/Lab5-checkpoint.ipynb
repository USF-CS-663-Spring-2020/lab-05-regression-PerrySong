# General strategy
1. Plot the data
2. Look at F-statistic, p-value
3. R**2
4. Estimated coefficients and their p-values
5. Exame X for homoscedasticity(sp?) using BF test

# Toluca data

```python
# Plot the predictor against the target
# Plot the residual against the target
# Determine the coefficients (slope, intercept) against the target

import pandas as pd

df = pd.read_csv('toluca.csv')
df
```

# Plot the predictor against the target

```python
import matplotlib.pyplot as plt
%matplotlib inline

df.plot(kind='scatter', x='lotSize', y='workHours')
```

# Plot the residual against the target

```python
import seaborn as sns
# A residual is the difference between the observed y-value (from scatter plot) 
# and the predicted y-value (from regression equation line). It is the vertical 
# distance from the actual plotted point to the point on the regression line. 
# You can think of a residual as how far the data "fall" from the regression line.

rl = sns.residplot('lotSize', 'workHours', data=df)
rl.set(xlabel='Lot Size', ylabel='Residual')
rl
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(df.drop('workHours', axis=1), df.iloc[:,1])
# reg.fit(df.iloc[:,0], df.iloc[:,1]) -- does not work. Why?
# Answer:
# df.iloc[:,1] is a one demensional array (serial), you can't use it as your 'x' in sklearn
# df.drop(...) -> pd.DataFrame
reg
```

# Determine the coefficients (slope, intercept) against the target

```python
print("coeficient = ", reg.coef_)
```

```python
print("intercept = ", reg.intercept_)
```

```python
import seaborn as sns
sns.lmplot(y='workHours', x='lotSize', data=df)
```

# F-statistic

H = X predicts Y
H0 = X ! predict Y (P-value: 0.05, 0.10)

```python
# 2. Look at F-statistic, p-value
import statsmodels.formula.api as smf

tolucaOLSmodel = smf.ols('workHours ~ lotSize', df).fit()
print('Fvalue', tolucaOLSmodel.fvalue, '\nPvalue:', tolucaOLSmodel.f_pvalue)
```

```python
# 3. R**2
tolucaOLSmodel.rsquared
```

```python
# 4. Slope and intercept
tolucaOLSmodel.params
```

```python
tolucaOLSmodel.pvalues
# We have more confident of intercept than lotSize
```

The intercept's p-value < 0.05, and the coefficient's p-value < 0.05, thus the result is suitable to the data

```python
# 5. BF Test
import scipy as sp

sample_1 = tolucaOLSmodel.resid[1:round(len(tolucaOLSmodel.resid)/2)]
sample_2 = tolucaOLSmodel.resid[round(len(tolucaOLSmodel.resid)/2) + 1:]
sp.stats.levene(sample_1, sample_2, center='median')
```

```python
sns.distplot(tolucaOLSmodel.resid)
```

```python
# Test whether this is a normal distribution
sp.stats.shapiro(tolucaOLSmodel.resid)[1]
```

```python
def plot_residual(xlabel, ylabel, df):
    rl = sns.residplot(xlabel, ylabel, data=df)
    rl.set(xlabel=xlabel, ylabel='Residual')
    return rl
```
